```
import hubspot
from pprint import pprint
from hubspot.crm.contacts import PublicGdprDeleteInput, ApiException

UPLOAD_API_TOKEN = dbutils.secrets.get(scope="key-vault-secrets",key="upload-api-accesstoken")
client = hubspot.Client.create(access_token=UPLOAD_API_TOKEN)
def delete_gdpr(id):
    public_gdpr_delete_input = PublicGdprDeleteInput(object_id=id)
    try:
        api_response = client.crm.contacts.gdpr_api.purge(public_gdpr_delete_input=public_gdpr_delete_input)
        pprint(api_response)
    except ApiException as e:
        print("Exception when calling gdpr_api->purge: %s\n" % e)
```

```
file_path = '/dbfs/mnt/hubspotarchivedids/archived_ids_01.json'
with open(file_path, 'r') as json_file:
    data = json.load(json_file)

print(len(data))
merged_list = [item for sublist in data for item in sublist]
print(len(merged_list))
```
